```python
from paddleocr import PaddleOCR, draw_ocr

class TableNormalizer:
    def __init__(self):
        # Initialize the PaddleOCR model
        self.ocr = PaddleOCR(use_angle_cls=True, lang='en')
    
    def extract_table(self, image_path,img_file):
        # Load image
        
        result = ocr.ocr(f"{image_path}/{img_file}", cls=False)
        # Use PaddleOCR to extract text
        # result = self.ocr.ocr(img, cls=True)
        return result

    def apply_post_processing(self, ocr_results, post_processors):
        # Apply each post-processing step to the OCR results
        for processor in post_processors:
            ocr_results = processor.process(ocr_results)
        return ocr_results

class PostProcessor:
    def process(self, data):
        # Define this in subclasses for specific post-processing
        raise NotImplementedError("This method should be overridden by subclasses.")

# Example of a specific PostProcessor
class RemoveEmptyCellsProcessor(PostProcessor):
    def process(self, data):
        # Removes empty cells from OCR results
        return [item for item in data if any(text.strip() for _, text in item[1])]

# Usage
# if __name__ == "__main__":
```
