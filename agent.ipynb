```python
import os
import json
from pathlib import Path
from pdfminer.pdfparser import PDFParser, PDFSyntaxError
from pdfminer.pdfdocument import PDFDocument, PDFNoOutlines
import json
import os
from dotenv import load_dotenv 
load_dotenv()
root_folder = os.environ["DEFAULT_LOCATION_FOR_TOPIC_COLLECTION"]
topics = r"Topics"
topics_folder = f"{root_folder}/{topics}"
input_file = "200R IM Input.pdf"
input_file_name = f"{os.environ['INPUT_FILES']}/{input_file}"
output_file_name = os.environ["OUTPUT_FILES"]
print(input_file_name, output_file_name)

def load_json(file_path):
   with open(file_path, 'r', encoding='utf-8') as file:
       return json.load(file)

def create_nested_dict(hierarchy_dict):
   def nested_insert(d, keys, value):
       for key in keys[:-1]:
           d = d.setdefault(key, {})
       d[keys[-1]] = value
       return d
   nested_dict = {}
   stack = []
   for key, level in hierarchy_dict.items():
       while len(stack) >= level:
           stack.pop()
       if stack:
           current_dict = nested_dict
           for parent in stack:
               current_dict = current_dict[parent]
           current_dict[key] = {}
           stack.append(key)
       else:
           nested_dict[key] = {}
           stack.append(key)
   return nested_dict

def create_folders_from_json(json_data, root_folder):
   def create_nested_folders(data, current_path):
       for key, value in data.items():
           new_path = os.path.join(current_path, key)
           os.makedirs(new_path, exist_ok=True)
           if isinstance(value, dict):
               create_nested_folders(value, new_path)
   if not os.path.exists(root_folder):
       os.makedirs(root_folder)
   create_nested_folders(json_data, root_folder)
   print("Folders created for Topics")


root_folder = os.environ["DEFAULT_LOCATION_FOR_TOPIC_COLLECTION"]
default_project = os.environ["DEFAULT_PROJECT"]
topics = r"topics"
topics_folder = f"{root_folder}/{default_project}/{topics}"
topics_folder, input_file_name
#create_folders_from_json(json_data, topics_folder):
## C:\Users\Ashish.Madkaikar\OneDrive - Emerson\GenaAI IMP\SSV_Marketing\200R\Topics
```

```python

from OCRManager import perform_ocr_on_images
from paddleocr import PaddleOCR,draw_ocr
ocr = PaddleOCR(use_angle_cls=True, lang='en')

topics_json_file_path = f"{topics_folder}\\topics.json"

def update_nested_dict(nested_dict, keys, value):
    """Update nested dictionary with OCR text."""
    for key in keys[:-1]:
        nested_dict = nested_dict.setdefault(key, {})
    nested_dict[keys[-1]] = value
def perform_ocr_on_images(root_dir, json_file_path):
    ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Initialize PaddleOCR

    # Load existing data from JSON file
    if os.path.exists(json_file_path):
        with open(json_file_path, 'r') as json_file:
            nested_dict = json.load(json_file)
    else:
        nested_dict = {}

    for subdir, _, files in os.walk(root_dir):
        # Calculate relative path from root directory and split it into parts
        relative_path = os.path.relpath(subdir, root_dir).split(os.sep)
        for file in files:
            if file.endswith('.png'):
                file_path = os.path.join(subdir, file)
                result = ocr.ocr(file_path, cls=True)
                ocr_text = '\n'.join([line[1][0] for line in result[0]])
                file_key = os.path.splitext(file)[0]
                txt_filename = f"{file_path.replace(".png",".txt")}"
                f = open(txt_filename, "w")
                f.write(ocr_text)
                f.close()
                update_nested_dict(nested_dict, relative_path + [file_key], ocr_text)
                print(f'OCR completed for {file_path}. Text updated in nested dictionary.')

    # Save updated data back to JSON file
    with open(json_file_path, 'w') as json_file:
        json.dump(nested_dict, json_file, indent=4)

    return nested_dict

ocr_data = perform_ocr_on_images(topics_folder, topics_json_file_path)
#print(ocr_data)
```

```python
import json
root_folder = os.environ["DEFAULT_LOCATION_FOR_TOPIC_COLLECTION"]
default_project = os.environ["DEFAULT_PROJECT"]
topics = r"topics"
topics_folder = f"{root_folder}/{default_project}/{topics}"
topics_folder, input_file_name
topics_json_file_path = f"{topics_folder}\\topics.json"
def extract_topics_and_content(data):
    extracted_data = []

    def recurse_extract(json_object, current_topic=""):
        if isinstance(json_object, dict):
            for key, value in json_object.items():
                if not key.isnumeric():
                    new_topic = f"{current_topic} {key}".strip()
                else:
                    new_topic = f"{current_topic} ".strip()
                #print(current_topic, key)
                recurse_extract(value, new_topic)
        elif isinstance(json_object, str):
            extracted_data.append({"topic": current_topic, "content": json_object})
        elif isinstance(json_object, list):
            for item in json_object:
                recurse_extract(item, current_topic)

    recurse_extract(data)
    return extracted_data

# Load the JSON data
with open(topics_json_file_path, 'r') as f:
    data = json.load(f)
    

# Convert the data
converted_data = extract_topics_and_content(data)

# Print the converted data
# Save updated data back to JSON file
# with open(json_file_path, 'w') as json_file:
#     json.dump(nested_dict, json_file, indent=4)
print(json.dumps(converted_data, indent=4))

```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained GPT-2 model and tokenizer
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
gpt2_model = GPT2LMHeadModel.from_pretrained('gpt2')

# Define a function to retrieve and generate responses
def generate_response(query, top_k=5):
    # Vectorize the query
    query_embedding = model.encode([query], convert_to_tensor=True).cpu().detach().numpy()

    # Retrieve the top_k most similar documents
    distances, indices = index.search(query_embedding, top_k)

    # Concatenate the retrieved documents
    retrieved_texts = " ".join([converted_data[idx] for idx in indices[0]])
    print(f"Retrieved context : {retrieved_texts}")
    # Generate a response using GPT-2
    input_ids = tokenizer.encode(query + " " + retrieved_texts, return_tensors='pt')
    output = gpt2_model.generate(input_ids, max_length=1024, num_return_sequences=1)

    return tokenizer.decode(output[0], skip_special_tokens=True)

# Example usage
query = "find information for Specifications"
response = generate_response(query)
print(response)
```
